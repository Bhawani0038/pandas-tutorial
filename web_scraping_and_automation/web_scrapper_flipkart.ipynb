```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url_list = []
brand_list = []
discription_list = []
price_list = []
discount_list = []


for n in range(1, 51):
    url = f"https://www.flipkart.com/search?q=shoes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={n}"
    code = requests.get(url)


    soup = BeautifulSoup(code.text)
    product_cards = soup.select('._1sdMkc')  # Each product widget/card

    for card in product_cards:
        # Extract image URL
        image = card.select_one('._53J4C-')
        url_list.append(image['src'] if image else None)

        # Extract brand
        brand = card.select_one('.syl9yP')
        brand_list.append(brand.text if brand else None)

        # Extract description
        description = card.select_one('.WKTcLC')
        discription_list.append(description.text if description else None)

        # Extract price
        price = card.select_one('.Nx9bqj')
        price_list.append(price.text if price else None)

        # Extract discount
        discount = card.select_one('.UkUFwK')
        discount_list.append(discount.text if discount else None)


shoes_data = pd.DataFrame(
    {
        "url"         : url_list, 
        "brand"       : brand_list, 
        "price"       : price_list, 
        'discount'    : discount_list, 
        'discription' : discription_list
    }
)

shoes_data.to_excel('shoes.xlsx')
```
