```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate data
np.random.seed(42)

study_hours = np.linspace(1, 10, 100)
# Negative correlation with added noise, and decreasing density
distractions = 20 / (study_hours + np.random.normal(0, 0.5, 100)) + np.random.normal(0, 1.5, 100)

# Reduce density of points as study_hours increase
mask = np.random.rand(100) < (1 / study_hours)
study_hours_sparse = study_hours[mask]
distractions_sparse = distractions[mask]

plt.figure(figsize=(10, 6))
plt.scatter(study_hours_sparse, distractions_sparse, color='tomato', edgecolor='black', alpha=0.8)

plt.title("Study Hours vs Distractions (Negative Correlation with Decreasing Density)")
plt.xlabel("Study Hours per Day")
plt.ylabel("Number of Distractions")
plt.grid(True)
plt.show()
```
